```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.DataFrame({'x':[2, 4, 10, 12, 3, 20, 30, 11, 25, 7]})
kmeans = KMeans(n_clusters = 2, max_iter = 100)
kmeans.fit(data)
print('\nSum of Square Error/ WCSS is :',kmeans.inertia_)
data['labels'] = kmeans.labels_
```

```python
k_values = []
wcss_values = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters = k, max_iter = 100)
    kmeans.fit(data)
    k_values.append(k)
    wcss_values.append(kmeans.inertia_)
    
plt.plot(k_values, wcss_values)
plt.show() 
```

```python
x_axis = np.array([1,2,3,18,19,20, 3, 2, 4])
y_axis = np.array([1,2,3,18,19,20, 16, 17, 19])

data = pd.DataFrame({'x':x_axis, 'y':y_axis})

data
```

```python
plt.figure(figsize=(5, 3))
plt.plot()
plt.title('Dataset')
plt.scatter(x_axis,y_axis)
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, max_iter = 100).fit(data)
print('\nSum of Square Error/ WCSS is :',kmeans.inertia_)

plt.figure(figsize=(5, 3))
plt.plot()
plt.title('Dataset')
plt.scatter(x_axis,y_axis, c=kmeans.labels_)
plt.show()
```

```python
for k in range(1, 10):
    kmeans = KMeans(n_clusters = k, max_iter = 100).fit(data)
    print('k-value - ', k, '\nSum of Square Error/ WCSS is :',kmeans.inertia_)
```
