## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S
# Replace missing values in 'Embarked' column with 'S'
df['Embarked'].fillna('S', inplace=True)

# Verify if missing values are replaced
print(df['Embarked'].isnull().sum())
```

```python
#Impute the missing values in Fare column with median value
# Replace missing values in 'Fare' column with its median
df['Fare'].fillna(df['Fare'].median(), inplace=True)

# Verify if missing values are replaced
print(df['Fare'].isnull().sum())  # Should print 0 if all missing values are replaced
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
import numpy as np

# Compute mean and standard deviation of the non-missing Age values
age_mean = df['Age'].mean()
age_std = df['Age'].std()

# Count of missing values
missing_count = df['Age'].isnull().sum()

# Generate random values from normal distribution
random_ages = np.random.normal(loc=age_mean, scale=age_std, size=missing_count)

# Fill missing values with these random ages
df.loc[df['Age'].isnull(), 'Age'] = random_ages

# Verify if missing values are replaced
print(df['Age'].isnull().sum())  # Should print 0
```

## Feature Engineering - Creating new features

```python
# Create and add a new feature FamilySize as a combination of SibSp and Parch
#FamilySize = Parch + SibSp + 1
# Create the FamilySize feature
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Display the first few rows to verify the new feature
print(df[['SibSp', 'Parch', 'FamilySize']].head())
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
# Create the IsAlone feature
df['IsAlone'] = df['FamilySize'] == 1  # This will create a boolean column (True/False)

# Convert boolean values to integers (0 = Not Alone, 1 = Alone) if needed
df['IsAlone'] = df['IsAlone'].astype(int)

# Display first few rows to verify
print(df[['FamilySize', 'IsAlone']].head())
```

## Model Building

```python
#Prepare X and y
# Define target variable (y)
y = df['Survived']  # The column we want to predict

# Define feature set (X)
X = df.drop(columns=['Survived'])  # Drop the target column from the dataset

# Display the first few rows of X and y
print("Features (X):")
print(X.head())

print("\nTarget (y):")
print(y.head())
```

```python
# Label encoding of categorical features
from sklearn.preprocessing import LabelEncoder

# List of categorical columns to encode
categorical_cols = ['Sex', 'Embarked']

# Apply Label Encoding
label_encoders = {}  # Dictionary to store encoders for later use (if needed)
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])  # Transform categorical values into numeric
    label_encoders[col] = le  # Store the encoder

# Display the first few rows to verify encoding
print(df.head())
```

```python
#Do a pairplot of all features
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pairplot of the dataset
sns.pairplot(df, diag_kind='kde', corner=True)  # Use KDE for diagonal & hide duplicate plots

# Show the plot
plt.show()
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
from sklearn.model_selection import train_test_split

# Splitting the dataset (90% Train, 10% Validation)
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)

# Display the sizes of the splits
print(f"Training set: X_train = {X_train.shape}, y_train = {y_train.shape}")
print(f"Validation set: X_validation = {X_validation.shape}, y_validation = {y_validation.shape}")
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Define X (features) and y (target)
X = df.drop(columns=['Survived'])  # Features
y = df['Survived']  # Target variable

# Split data (90% train, 10% validation)
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)

# Initialize and train Decision Tree model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_validation)

# Evaluate accuracy
accuracy = accuracy_score(y_validation, y_pred)
print(f"Validation Accuracy: {accuracy:.4f}")

```

```python
#Print the tree

# Importing required packages for visualization
from sklearn.tree import export_text, plot_tree
import matplotlib.pyplot as plt

# Print the textual representation of the tree
tree_rules = export_text(model, feature_names=list(X.columns))
print(tree_rules)

# Plot the tree
plt.figure(figsize=(15, 8))  # Set figure size
plot_tree(model, filled=True, feature_names=X.columns, class_names=['Not Survived', 'Survived'])
plt.show()

```

```python
# plotting tree with max_depth=3
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Train a new Decision Tree with max_depth=3
model_limited = DecisionTreeClassifier(max_depth=3, random_state=42)
model_limited.fit(X_train, y_train)

# Plot the tree with max_depth=3
plt.figure(figsize=(15, 8))  # Set figure size
plot_tree(model_limited, filled=True, feature_names=X.columns, class_names=['Not Survived', 'Survived'])
plt.show()
```
